```python
import pandas as pd 

attr = pd.read_csv("../datasets/attribute_set/in_shop_attr/list_eval_partition.txt", delim_whitespace=True, header=None, names=["path", "id", "group"]) 
attr.head()    
```

```python
attr.head() 
```

```python
base_path = "../datasets/big_ds/img-in_shop/img/" 
base_path 
```

```python
attr.path = attr.path.apply(lambda x: base_path + x) 
attr.tail()
```

```python
col_names = pd.read_fwf("../datasets/attribute_set/in_shop_attr/list_attr_cloth.txt", delim_whitespace=True).to_numpy().squeeze() 
i = 0
for col in col_names:
    col_names[i] = str(i) + "_" + col 
    i = i + 1 
col_names[:5], len(col_names)
```

```python
len(set(col_names))
```

```python
labels = pd.read_csv("../datasets/attribute_set/in_shop_attr/list_attr_items.txt", header=None, delim_whitespace=True, names=["id"] + list(col_names))
labels.head() 
```

```python
items = [] 
for col in labels.columns[1:]:
    for item in labels[col].value_counts().index.values: 
        items.append(item) 

set(items) 
```

```python
for col in range(1, len(labels.columns)):
    labels.iloc[:, col] = labels.iloc[:, col].astype('category').cat.codes.values 
    if col % 100 == 0: 
        print(col) 
```

```python
labels.head() 
```

```python
items_one_label = [] 
for col in labels.columns[1:]:
    if len(labels[col].value_counts().index.values) == 1:
        print(col, end=" ")
        items_one_label.append(col)   

len(items_one_label)
```

```python
labels.shape
```

```python
labels.drop(items_one_label, axis=1, inplace=True)
```

```python
labels.head() 
```

```python
labels.shape
```

```python
data = pd.merge(attr, labels, how="inner", on="id")
data.head() 
```

```python
data.to_csv("../datasets/attribute_set/in_shop_attr/in_shop_data_all.csv") 
```
