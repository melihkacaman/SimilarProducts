# Data Preparation For Fine Grained Attributes 

```python
import pandas as pd 
import numpy as np 

paths_df = pd.read_csv("../datasets/attribute_set/train_fine_grained.txt", names=["paths"], sep=" ", header=None) 
```

```python
paths_df.info() 
```

```python
columns = pd.read_csv("../datasets/attribute_set/fine_grained_column_names.txt", header=None, delimiter=" ").to_numpy().squeeze()
```

```python
attr_df = pd.read_csv("../datasets/attribute_set/train_attr_fine_grained.txt", header=None, delimiter=" ").iloc[:, :-1]
attr_df.columns = columns
attr_df.head() 
```

```python
attr_df.shape, paths_df.shape 
```

```python
base = "../datasets/big_ds/img-001/" 
base
```

```python
paths_df = paths_df.apply(lambda x: base + x) 
paths_df.head() 
```

```python
data = pd.concat([paths_df, attr_df], axis=1)
data.head() 
```

```python
val_attr_df = pd.read_csv("../datasets/attribute_set/val_attr_fine_grained.txt", header=None, delimiter=" ").iloc[:, :-1]
val_attr_df.columns = columns
val_attr_df.head() 
```

```python
val_paths_df = pd.read_csv("../datasets/attribute_set/val_fine_grained.txt", names=["paths"], sep=" ", header=None)
val_paths_df = val_paths_df.apply(lambda x: base + x) 
val_paths_df.head() 
```

```python
val_data = pd.concat([val_paths_df, val_attr_df], axis=1)
val_data.head() 
```

```python
val_data.shape, data.shape 
```

```python
TRAIN_SIZE = 14000 

data = pd.concat([data, val_data], axis=0) 
```
